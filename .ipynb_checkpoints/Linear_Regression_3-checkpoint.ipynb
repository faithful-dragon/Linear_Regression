```python
X_data = data.drop('Price in INR', axis=1)
Y_data = data['Price in INR']
```

```python
def remove_rupee_sign(data):
    return [item.replace('₹', '') for item in data]
Y_data = remove_rupee_sign(Y_data)
```

```python
Y_data = pd.DataFrame(Y_data, columns=["Price"])
```

```python
Y_data.head()
```

```python
Y_data['Price']=Y_data['Price'].str.replace(',','')
```

```python
Y_data
```

```python
Y_data = Y_data['Price'].astype('float64')
```

```python
Y_data
```

```python
# Label Encoding
lst=["Phone Name","Rating ?/5" ,"Number of Ratings","RAM","ROM/Storage","Back/Rare Camera","Front Camera" ,"Battery" ,"Processor" ,"Date of Scraping"]
le=LabelEncoder()
for i in lst:
    X_data[i]=le.fit_transform(X_data[i])
```

```python
X_data.head()
```

```python
Y_data
```

```python
X_train , X_test, Y_train , Y_test = train_test_split(X_data, Y_data, test_size =0.3 , random_state = 42)
```

```python
print("X_train shape :", X_train.shape)
print("y_train shape :", Y_train.shape)
print("X_test shape :", X_test.shape)
print("y_test shape :", Y_test.shape)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_test_pred = lr.predict(X_test)
Y_train_pred = lr.predict(X_train)
```

```python
Y_test_pred
```

```python
from sklearn.metrics import r2_score
```

```python
print("R^2 on training  data ",r2_score(Y_train_pred, Y_train))
print("R^2 on testing data ",r2_score(Y_test_pred,Y_test))  # bad prediction
# more R2 value means more accurate model
```
